```python
# load library
import pandas as pd
# create dataframe
dataframe = pd.DataFrame()
# add columns
dataframe['Name'] = ['Lemon','Apple','Banana']
dataframe['Price'] = [5000,8000,3000]
dataframe['License'] = [False,True,True]
# Show data
dataframe
```

```python
dataframe.shape 
```

```python
# df la datafame
df = dataframe
df.info()
```

```python
# print column name
list(df.columns)
```

```python
df.describe()
#count: so luong 
#mean: gia tri trung binh
#std: do lech chuan
```

```python
# loc iloc (location  & index location)
df.loc[(df.Price == 5000)]
```

```python
df.loc[(df.Price == 8000) | (df.Name == "Banana")]  # | : or 
```

```python
df.loc[(df.Price == 5000) & (df.Name == "Lemon")]  # & : and
```
