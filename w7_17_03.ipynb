```python
import pandas as pd 
url = '/50_Startups.csv'
data = pd.read_csv(url)
data.head()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
sns.pairplot(data)
```

```python
data.plot(kind = 'scatter', x='R&D Spend', y='Profit', figsize=(9,9), color = 'blue')

correlation = data.corr() 
print(correlation)
```

```python
sns.heatmap(correlation,annot=True)
 
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['R&D Spend']),y=data['Profit'])
print('Gia tri Intercept la:',model.intercept_)
print('Gia tri coeffient la :',model.coef_)
```

```python
X=pd.DataFrame(data['R&D Spend'])
y=data['Profit']
model.score(X,y)
```

```python
data.plot(kind='scatter',x='R&D Spend',y='Profit',figsize=(9,9),color='black')
y_pred=model.predict(X)
plt.plot(data['R&D Spend'],y_pred,color='blue')
```

```python
model = linear_model.LinearRegression() 
X = pd.DataFrame([data['R&D Spend'], data['R&D Spend']**2, data['R&D Spend']**3, data['R&D Spend']**4]).T
y = data['Profit']
model.fit(X,y)
print('Intercept:', model.intercept_)
print('Coeffient:', model.coef_)
print('Score:', model.score(X,y))
range = np.arange(1.5,5.5,0.1)
X_range = pd.DataFrame([range,range**2,range**3,range**4]).T
y_pred = model.predict(X_range)
data.plot(kind='scatter', x='R&D Spend', y='Profit', figsize=(9,9), color='green', xlim=(0,7))
plt.plot(range,y_pred,color='brown')
```
